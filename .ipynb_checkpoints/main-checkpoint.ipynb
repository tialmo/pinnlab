```python
#main.ipynb
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from netgen.meshing import ImportMesh
import meshio
from scipy.interpolate import griddata
from matplotlib.ticker import MultipleLocator

import importlib
import sample
importlib.reload(sample)

import evaluate
importlib.reload(evaluate)

import models
importlib.reload(models)

import lapinn
importlib.reload(lapinn)

shaft_radius = 0.0325 # 0.65 mm with diameter beeing 1.3 mm

domain_bounds = {'x': [-1, 1], 'y': [-1, 1], 'z': [-1, 1]}
pde_params = {
    'contact1_cylinder_radius': shaft_radius,
    'contact1_cylinder_height': 0.075,
    'contact1_cylinder_center': [0, 0, shaft_radius+0.055],
    'contact1_value': 1.0,

    'contact2_cylinder_radius': shaft_radius,
    'contact2_cylinder_height': 0.075,
    'contact2_cylinder_center': [0, 0, 0.1375+0.055],
    'contact2_value': 0.0,

    'neumann_cylinder_radius': shaft_radius,
    'neumann_cylinder_height': 1, 
    'neumann_cylinder_center': [0.0, 0.0, 0.5], 
    'neumann_value': 0.0,
}

batch = sample.sample_bipolar_geometry_points(
    domain_bounds=domain_bounds,
    pde_params=pde_params,
    n_r=20000,
    n_cube_neumann=2400,
    n_shaft_neumann=1200,
    n_contact1_dirichlet=600,
    n_contact2_dirichlet=600,
    seed=42
)

fig = sample.plot_sampled_points(batch, max_points=10000, title="Bipolar Geometry Sampling Check")
print("Sampled.")

batch = lapinn.override_collocation_from_npz(
    batch=batch,
    npz_path="remapped_scaled_data.npz",
    pde_params=pde_params,
    sigma_scale=1e-3,
    filter_cylinders=True
)


 # ---- build model ----
#net = models.build_mlp(in_dim=3, hidden=(128, 128, 128, 128), out_dim=1, act="tanh") #basic pinn
#pinn = models.BasicLaplacePINN(net)
#opt = tf.keras.optimizers.Adam(1e-3)

w = 120
la = lapinn.LaplaceLAPINN(
    layers_main=[3, w, w, w, w, w, 1],
    layers_lan=[1, w, w, w, w, w, 1],
    pde_params=pde_params
)

la.prepare_interface_from_batch(batch)


epochs = 20000
for e in range(epochs):
    total, lr, lif, ln, lc1, lc2 = la.train_step_on_batch(batch)
    if e % 100 == 0:
        print(f"epoch {e:5d} | total={total.numpy():.3e} "
              f"| r={lr.numpy():.3e} if={lif.numpy():.3e} "
              f"n={ln.numpy():.3e} c1={lc1.numpy():.3e} c2={lc2.numpy():.3e}")
print("Done.")


evaluate.compare_with_fem(la, shaft_radius)
```
